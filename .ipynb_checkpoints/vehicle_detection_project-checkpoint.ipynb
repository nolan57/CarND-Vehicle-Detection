```python
def topleft(w, h, xstart, ystart, window_size, scale, overlap):
    
    y  = ystart
    x  = xstart
    
    topleftx = []
    topleftyx = {}
    topleftyx[y] = topleftx
    
    while 1:
        
        if y + window_size < h:
        
            while 1:
                if x + window_size > w:
                    x = xstart
                    topleftx = []
                    break                
                topleftx.append(x)
                topleftyx[y] = topleftx
                x = x + int(window_size*overlap)
                    
            y = y + int(window_size*overlap)
            window_size = int(window_size*scale)
        else:
            break
                                 
    return topleftyx
```

```python
w = 1280
h = 650
xstart = 0
ystart = 400
window_size = 64
scale = 1
overlap = 0.5

topleftyx = topleft(w, h, xstart, ystart, window_size, scale, overlap)
```

```python
for toplefty in topleftyx.keys():
    
    print(toplefty)
    print(topleftyx[toplefty])
```

```python
def mult_scale_windows(topleftyx, overlap):
    
    msw = []
    
    for toplefty in topleftyx.keys():
        
        if overlap == 1:
            window_size = topleftyx[toplefty][1] - topleftyx[toplefty][0]
        else:
            window_size = topleftyx[toplefty][2] - topleftyx[toplefty][0]
        for toplefx in topleftyx[toplefty]:
            
            buttomrightx = toplefx + window_size
            buttomrighty = toplefty + window_size
            
            msw.append(((toplefx, toplefty), (buttomrightx, buttomrighty)))
    
    return msw
```

```python
windows  = mult_scale_windows(topleftyx, overlap)

for window in windows:
    
    print(window)
```
